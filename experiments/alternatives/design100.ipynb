```python
import pandas as pd
import numpy as np
from pathlib import Path
import sys

for parent in [Path.cwd(), *Path.cwd().parents]:
    if (parent / "config" / "config.yaml").exists():
        if str(parent) not in sys.path:
            sys.path.insert(0, str(parent))
        break

from config import get_config

cfg = get_config()
attrs = cfg.get_attributes()  

attributes = list(attrs.keys())
attribute = attributes 
levels = {k: v.get("values", []) for k, v in attrs.items()}
df = pd.read_csv("./Design.csv").rename(columns={"Unnamed: 0":"real_profile_id"})
for j, i in enumerate(levels.keys()):
    if len(levels[i])==df[f'A{j+1}'].max():
        df.rename(columns={f'A{j+1}':i}, inplace=True)
        df[i] = df[i].map(lambda x: levels[i][x-1])
#suffix
df['real_profile_id'] = df['real_profile_id'].apply(lambda x: f'r{x}')
df.to_csv("./design100.csv",index=False)
df.head()
```
