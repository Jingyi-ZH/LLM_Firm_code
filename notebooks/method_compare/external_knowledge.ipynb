```python
for real_profile_id in ['iPhone 17','iPhone 17 Air','iPhone 17 Pro','iPhone 17 Pro Max']:
    df_rag = pd.read_csv(f"../output/rag_langchain_{real_profile_id}_fixreal500.csv")
    df_longinjection = pd.read_csv(f"../output/re16ru17_{real_profile_id}_fixreal5000.csv")
    df_injection = pd.read_csv(f"../output/reali16_{real_profile_id}_fixreal5000.csv")
    df_injection = df_injection[df_injection['profile_id'].isin(sampled_ids)]
    df_longinjection = df_longinjection[df_longinjection['profile_id'].isin(sampled_ids)]
    df_rag = df_rag[df_rag['profile_id'].isin(sampled_ids)]
    assert len(df_longinjection)==len(df_injection)
    longinjection_count = np.sum(df_longinjection['chosen_id']==real_profile_id)
    injection_count = np.sum(df_injection['chosen_id']==real_profile_id)
    rag_count = np.sum(df_rag['rag_chosen_id']==real_profile_id)
    print(f"{real_profile_id} & {injection_count/50}\% & {longinjection_count/50}\% & {rag_count/5}\% \\\\")
```
